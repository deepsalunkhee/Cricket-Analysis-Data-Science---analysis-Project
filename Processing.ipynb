```python
match_ids_dict={}

for index, row in df_match.iterrows():
    key1=row['team1']+' Vs '+row['team2']
    key2=row['team2']+' Vs '+row['team1']
    
    match_ids_dict[key1]=row['match_id']
    match_ids_dict[key2]=row['match_id']
    
match_ids_dict
```

This code is creating a Python dictionary named match_ids_dict that maps a match key to a match ID for a given DataFrame df_match.

For each row in the df_match DataFrame, the code is concatenating the names of the two teams playing in a particular match, separated by the string "Vs", to create two possible match keys: key1 and key2. This is done because it is not clear which team will be listed first in the DataFrame for each match.

Next, the code is assigning the match ID (which is stored in the 'match_id' column of df_match) to both key1 and key2 in the match_ids_dict dictionary. This allows the user to easily retrieve the match ID by using either key1 or key2, regardless of the order in which the teams were listed in the original DataFrame.

In summary, this code is creating a dictionary that can be used to quickly look up the match ID for a given match by using either the team1VsTeam2 or team2VsTeam1 key.

```python
df_batting["match_id"]= df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('batting.csv',index=False)
```

Repeating the same Process fro match summery=>

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv('summary.csv', index = False)
```
